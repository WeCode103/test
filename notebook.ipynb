```python
def show_range(numbers):
    """Return True if all numbers are between 0 and 100 (inclusive), else False."""
    return all(0 <= n <= 100 for n in numbers)

print(show_range([0, 44, 100]))   # True
print(show_range([-1, 20,101]))    # False


```
